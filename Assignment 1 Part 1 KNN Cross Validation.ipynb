```python
import numpy as np
import matplotlib.pyplot as plt
import KNN as NN
import LOAD_CIFAR as ld
from sklearn.utils import shuffle
```

```python
def sample(train_data, train_label):
    sorted_indices = train_label.argsort()
    train_label.sort()
    sorted_train = np.array(train_data[sorted_indices])
    
    for i in range(1, 11):
        if i == 1:
            initial = 5000*(i-1)
            end = initial + 1000
            new_train = sorted_train[initial:end,]
            new_label = train_label[initial:end,]
        else:
            initial = 5000*(i-1)
            end = initial + 1000
            new_train = np.append(new_train, sorted_train[initial:end],axis = 0)
            new_label = np.append(new_label, train_label[initial:end])
    return new_train, new_label
```

```python
def divide(train_data, train_label,i):
    if i==0:
        return train_data[0:6666,] , train_label[0:6666], train_data[6666:10000,] , train_label[6666:10000]
    elif i==1:
        return np.append(train_data[0:3333,], train_data[6667:10000,],axis = 0) , \
                np.append(train_label[0:3333], train_label[6667:10000]),\
                train_data[3333:6667,] , train_label[3333:6667]
    else:
        return train_data[3334:10000,] , train_label[3334:10000], train_data[0:3334,] , train_label[0:3334]
```

```python
data_dir = 'cifar-10-python'

X_train, y_train = ld.load_cifar10_cv(data_dir)

X_train = np.reshape(X_train, (X_train.shape[0], -1))

X_train, y_train = sample(X_train,y_train)

y_train, X_train = shuffle(y_train, X_train, random_state=0)
```

```python
k = [1, 5,  7,  10, 20, 50, 100]
accuracy = []
accurcayperk = []

for x in k:
    for i in range(3):
        X_datatrain, y_labeltrain, X_test, y_test = divide(X_train, y_train,i)
        classifier = NN.KNN()
        classifier.train(X_datatrain, y_labeltrain)
        y_test_pred = classifier.predict(X_test,'L1',x)
        num_correct = np.sum(y_test_pred == y_test)
        acc= float(num_correct) /(3334)
        print(acc)
        accurcayperk.append(acc)
    accuracy.append(list(accurcayperk))
        
        
print(accurcayperk)
```
