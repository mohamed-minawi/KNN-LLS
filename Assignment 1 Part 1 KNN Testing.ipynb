```python
import numpy as np
import matplotlib.pyplot as plt
import LOAD_CIFAR as ld
import KNN as NN
from sklearn.utils import shuffle
```

```python
def sample(train_data, train_label):
    sorted_indices = train_label.argsort()
    train_label.sort()
    sorted_train = np.array(train_data[sorted_indices])
    
    for i in range(1, 11):
        if i == 1:
            initial = 5000*(i-1)
            end = initial + 1000
            new_train = sorted_train[initial:end,]
            new_label = train_label[initial:end,]
        else:
            initial = 5000*(i-1)
            end = initial + 1000
            new_train = np.append(new_train, sorted_train[initial:end],axis = 0)
            new_label = np.append(new_label, train_label[initial:end])
    return new_train, new_label
```

```python
data_dir = './cifar-10-python/'

X_train, y_train, X_test, y_test = ld.load_cifar10_data(data_dir)

X_train, y_train = sample(X_train,y_train)

y_train, X_train = shuffle(y_train, X_train, random_state=0)

X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_test = np.reshape(X_test, (X_train.shape[0], -1))
```

```python
# Create a NN classifier instance and train it with training data
classifier = NN.KNN()
classifier.train(X_train, y_train)
```

```python
y_test_pred = classifier.predict(X_test,'L1',10)
```

```python
num_correct = np.sum(y_test_pred == y_test)
num_incorrect = 10000 - num_correct
accuracy = float(num_correct) /(10000)
print(num_correct,num_incorrect,10000,accuracy)
for i in range(10):
    print(i,  np.sum( np.logical_and(y_test_pred == y_test, y_test == i)) / float(np.sum(y_test == i)) )
```
